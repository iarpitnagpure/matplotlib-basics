```python
import matplotlib.pyplot as plt
import numpy as np

# random data for plotting
# np.random.rand(d0, d1, ..., dn): Generates an array of given shape and populates it with random samples from a uniform distribution over [0, 1).
x = np.random.rand(50) * 10                             # Random data for x-axis
y = np.random.rand(50) * 100                            # Random data for y-axis
print(x)

# Scatter plot
plt.scatter(x, y)  # Scatter plot with blue circles
plt.show()
```

```python
# styling the scatter plot
x = np.random.rand(500) * 10                                # Random data for x-axis
y = np.random.rand(500) * 100                               # Random data for y-axis
# plt.scatter(x, y), color='color', marker='marker_style, s=size, alpha=transparency (0 to 1)                                 
plt.scatter(x, y, color='blue', marker='D', s=40, alpha=0.5) 

plt.title("Student Marks Analysis", fontsize=14)            # Title with font size 14
plt.grid(True)                                              # Show grid (Property available for all plots)
plt.xlabel("Number of Hours Studied", fontsize=12)          # X-axis label with font size 12
plt.ylabel("Marks Obtained", fontsize=12)                   # Y-axis label with font size 12
plt.legend(["Student Marks"], loc='upper right')             # Legend at upper right position
plt.show()
```

```python
# Bar plot
subjects = ['Maths', 'Science', 'English', 'History', 'Geography']
marks = [88, 92, 85, 90, 87]

plt.bar(subjects, marks, color='m')  # Bar plot with magenta bars
plt.show()
```

```python
# Bar plot with customizations
# Bar plot with red bars, black edges, and width 0.5, with hatch pattern
plt.bar(subjects, marks, color='red', edgecolor='black', width=0.5, hatch='//')     # Bar plot with red bars, black edges, and width 0.5, with hatch pattern
plt.title("Student Marks in Different Subjects", fontsize=14)                       # Title with font size 14
plt.xlabel("Subjects", fontsize=12)                                                 # X-axis label with font size 12
plt.ylabel("Marks Obtained", fontsize=12)                                           # Y-axis label with font size 12

plt.show()
```

```python
# Bar plot customization with different Hatches patterns
hatches = ['/', '\\', '|', '-', 'x']
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Create bar plot and store the bar containers in 'bars'
bars = plt.bar(subjects, marks) 
plt.title("Student Marks in Different Subjects", fontsize=14)                       # Title with font size 14
plt.xlabel("Subjects", fontsize=12)                                                 # X-axis label with font size 12
plt.ylabel("Marks Obtained", fontsize=12)                                           # Y-axis label with font size 12

# For loop to apply different hatch patterns, color, and edge color to each bar
for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)                                                            # Apply different hatch patterns to each bar
    bar.set_color(colors[hatches.index(hatch)])                                     # Apply different colors to each bar
    bar.set_edgecolor('black')
plt.show()
```

```python
# Histogram plot
age = [22, 13, 6, 8, 10, 12, 15, 18, 97, 100, 95, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Histogram
plt.hist(age)
plt.show()
```

```python
# Histogram with customization
plt.hist(age, bins=8, color='c', edgecolor='black', alpha=0.7)      # Histogram with 8 bins, cyan color, black edges, and transparency
plt.title("Age Distribution of Students", fontsize=14)              # Title with font size
plt.xlabel("Age Groups", fontsize=12)                               # X-axis label with font size 12
plt.ylabel("Number of Students", fontsize=12)                       # Y-axis label with font size
plt.show()
```

```python
# Pie chart
labels = ['Maths', 'Science', 'English', 'History', 'Geography']
marks = [60, 92, 73, 80, 75]

plt.pie(marks, labels=labels)
plt.show()
```

```python
# Pie Chart with customizations
# Pie chart with percentage display, start angle, and custom colors, shadow effect, and explode effect (For science subject)
plt.pie(marks, labels=labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple', 'yellow'], shadow=True, explode=(0, 0.1, 0, 0, 0))
plt.title("Marks Distribution Across Subjects", fontsize=14)        # Title with font size 14
plt.show()
```

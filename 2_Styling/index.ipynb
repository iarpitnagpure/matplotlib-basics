```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [70, 80, 95, 99]

# Plotting the points
plt.plot(x, y, label="DSA Marks")

# Adding title to the plot
plt.title("Student Marks Analysis")

# Adding labels on x-axis
plt.xlabel("Number of Hours Studied")

# Adding labels on y-axis
plt.ylabel("Marks Obtained")

# Showing legend with location
plt.legend(loc="lower right")

plt.show()
```

```python
# Customizing the plot with colors, markers, and line styles along with multiple lines
x = [1, 2, 3, 4]
dsa_marks = [70, 80, 95, 99]
ml_marks = [50, 60, 65, 70]

# Plotting multiple lines on the same plot
# Plotting the points for DSA Marks with color orange, marker as circle, line style as dashed
plt.plot(x, dsa_marks, label="DSA Marks", color='orange', marker='o', linestyle='--')
# Plotting the points for ML Marks with color blue, marker as circle, line style as dashed
plt.plot(x, ml_marks, label="ML Marks", color='blue', marker='o', linestyle='--')

# Adding title to the plot with font properties
plt.title("Student Marks Analysis", fontdict={'fontweight': 'bold', 'fontsize': 14})

# Adding labels on x-axis
plt.xlabel("Number of Hours Studied", fontsize=12)

# Adding labels on y-axis
plt.ylabel("Marks Obtained", fontsize=12)

# Showing legend with location
plt.legend(loc="lower right")

# Customizing x and y axis ticks, This add ticks at specified positions on x and y axis
plt.xticks([1, 2, 3, 4])
plt.yticks([40, 60, 80, 100])

plt.show()
```

```python
x = [1, 2, 3, 4]
dsa_marks = [70, 80, 95, 99]
ml_marks = [50, 60, 65, 70]

# shorter syntax for color, marker and line style
plt.plot(x, dsa_marks, 'r--*', label="DSA Marks")  # 'r--*' means red color with dashed line and star markers
plt.plot(x, ml_marks, 'g-o', label="ML Marks")    # 'g-o' means green color with circle markers and solid line

plt.title("Student Marks Analysis", fontdict={'fontweight': 'bold', 'fontsize': 14})
plt.xlabel("Number of Hours Studied", fontsize=12)
plt.ylabel("Marks Obtained", fontsize=12)
plt.legend(loc="lower right")


# Save function saves the current plot to a file
# dpi (Optional) is dots per inch, higher the dpi, better the quality of the image
# transparent (Optional) if True, the background of the saved figure will be transparent
plt.savefig("student_marks_analysis.png")                               # Save the plot as a PNG file
plt.savefig("student_marks_analysis1.png", dpi=1000, transparent=True)  # Save the plot as a PNG file

# Show function clears the previous plot and displays the current plot
plt.show()
```

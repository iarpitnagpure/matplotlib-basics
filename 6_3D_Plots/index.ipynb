```python
import matplotlib.pyplot as plt
import numpy as np

# Create random data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 3D scatter plot
ax = plt.axes(projection='3d')                                      # Create a 3D axes object
ax.scatter(x, y, z, color='blue')                                   # Plot a 3D scatter plot
ax.set_title("3D Scatter Plot", fontsize=14)                        # Set title with font size 14
ax.set_xlabel("X Axis", fontsize=12)                                # Set x-axis label with font
ax.set_ylabel("Y Axis", fontsize=12)                                # Set y-axis label with font
ax.set_zlabel("Z Axis", fontsize=12)                                # Set z-axis label with font
plt.show()
```

```python
# 3D line plot
ax = plt.axes(projection='3d')                                          # Create a 3D axes object

x = np.arange(0, 50, 0.1)                                               # X values from 0 to 50 with step 0.1
y = np.sin(x)                                                           # Y values as sine of X        
z = np.cos(x)                                                           # Z values as cosine of X

# plotting the 3D line plot
ax.plot(x, y, z)                                                        # Plot a 3D line plot with green line and circle markers
ax.set_title("3D Line Plot", fontsize=14)                               # Set title with font size 14
ax.set_xlabel("X Axis", fontsize=12)                                    # Set x-axis label with font
ax.set_ylabel("Y Axis", fontsize=12)                                    # Set y-axis label with font
ax.set_zlabel("Z Axis", fontsize=12)                                    # Set z-axis label with font
plt.show()
```

```python
# 3D surface plot
ax = plt.axes(projection='3d')                                          # Create a 3D axes object

x = [1, 1, 2, 2]                                                         # X coordinates of the bars
y = [1, 2, 1, 2]                                                         # Y coordinates of the bars
z = [0, 0, 0, 0]                                                         # Z coordinates of the bars (base)
dx = [1, 1, 1, 1]                                                        # Width of the bars in x direction
dy = [1, 1, 1, 1]                                                        # Width of the bars in y direction
dz = [1, 2, 3, 4]                                                        # Height of the bars   

# Imagine x and y as the coordinates of the bottom left corner of each bar
# dx and dy as the width and depth of each bar
# dz as the height of each bar

# The 3D bar plot (x, y, z, dx, dy, dz)
# plotting the 3D bar plot
ax.bar3d(x, y, z, dx, dy, dz)                                           # Plot a 3D bar plot
ax.set_title("3D Bar Plot", fontsize=14)                                # Set title with font size 14
ax.set_xlabel("X Axis", fontsize=12)                                    # Set x-axis label with font
ax.set_ylabel("Y Axis", fontsize=12)                                    # Set y-axis label with font
ax.set_zlabel("Z Axis", fontsize=12)                                    # Set z-axis label with font
plt.show()
```

```python
# 3D surface plot
ax = plt.axes(projection='3d')                                               # Create a 3D axes object

# Generating random data for bars
x = np.random.randint(1, 10, 5)                                              # X coordinates of the bars
y = np.random.randint(1, 10, 5)                                              # Y coordinates of the bars
z = np.zeros(5)                                                              # Z coordinates of the bars (base)
dx = np.ones(1) * 0.5                                                        # Width of the bars in x direction
dy = np.ones(1)                                                              # Width of the bars in y direction
dz = np.random.randint(1, 10, 5)                                             # Height of the bars

print(x, y, z, dx, dy, dz)

# plotting the 3D bar plot
ax.bar3d(x, y, z, dx, dy, dz) 
ax.set_title("3D Bar Plot", fontsize=14)                                     # Set title with font size 14
ax.set_xlabel("X Axis", fontsize=12)                                         # Set x-axis label with font
ax.set_ylabel("Y Axis", fontsize=12)                                         # Set y-axis label with font
ax.set_zlabel("Z Axis", fontsize=12)                                         # Set z-axis label with font
plt.show()
```

```python
# Plot surface plot
# 3D surface plot
ax = plt.axes(projection='3d')                                               # Create a 3D axes object

# Genrerating data
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
# Meshgrid to create a grid of x and y values 
x, y = np.meshgrid(x, y)
print('X, Y:', x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# plotting the 3D surface plot with plasma colormap
ax.plot_surface(x, y, z, cmap='plasma')                                       # Plot a 3D surface plot
plt.show()  
```

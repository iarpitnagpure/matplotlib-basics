```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an array of 100 linearly spaced numbers from 0 to 10 using numpy
# linspace(start, stop, num): Generates num evenly spaced samples from start to stop (inclusive)
x = np.linspace(1, 10, 100)                             # 100 linearly spaced numbers from 0 to 10
print(x)                                                # Print the array of numbers

# subplots(nrows, ncols): Creates a grid of subplots with nrows rows and ncols columns
fig, axes = plt.subplots(2, 3)

# Plot different mathematical functions on each subplot with yellow dashed lines
# 'y--' specifies a yellow dashed line style
axes[0, 0].plot(x, np.sin(x), 'y--')                    # Plot sine wave in the first subplot with yellow dashed line
axes[0, 1].plot(x, np.cos(x), 'y--')                    # Plot cosine wave in the second subplot with yellow dashed line
axes[0, 2].plot(x, np.tan(x), 'y--')                    # Plot tangent wave in the third subplot with yellow dashed line
axes[1, 0].plot(x, np.sqrt(x), 'y--')                   # Plot square root in the fourth subplot with yellow dashed line
axes[1, 1].plot(x, np.log(x), 'y--')                    # Plot natural logarithm in the fifth subplot with yellow dashed line
axes[1, 2].plot(x, np.exp(x), 'y--')                    # Plot exponential in the sixth subplot with yellow dashed line

plt.tight_layout()                          # Adjust subplot parameters to give specified padding
plt.show()                                  # Display the plots
```

```python
# Generate an array of 100 linearly spaced numbers from 0 to 10 using numpy
# linspace(start, stop, num): Generates num evenly spaced samples from start to stop (inclusive)
x = np.linspace(1, 10, 100)                     # 100 linearly spaced numbers from 0 to 10

# subplots(nrows, ncols): Creates a grid of subplots with nrows rows and ncols columns
fig, axes = plt.subplots(2, 3)

# Plot different mathematical functions on each subplot with colored dashed lines
axes[0, 0].plot(x, np.sin(x), 'r--')                        # Plot sine wave in the first subplot with red dashed line
axes[0, 0].set_title("Sine Wave", fontsize=10)              # Set title for the first subplot with font size 10
axes[0, 0].set_ylabel("Sine Values", fontsize=8)            # Set y-axis label for the first subplot with font size 8
axes[0, 0].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the first subplot with font size 8


axes[0, 1].plot(x, np.cos(x), 'g--')                        # Plot cosine wave in the second subplot with green dashed line
axes[0, 1].set_title("Cosine Wave", fontsize=10)            # Set title for the second subplot with font size 10
axes[0, 1].set_ylabel("Cosine Values", fontsize=8)          # Set y-axis label for the second subplot with font size 8
axes[0, 1].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the second subplot with font size 8


axes[0, 2].plot(x, np.tan(x), 'b--')                        # Plot tangent wave in the third subplot with blue dashed line
axes[0, 2].set_title("Tangent Wave", fontsize=10)           # Set title for the third subplot with font size 10
axes[0, 2].set_ylabel("Tangent Values", fontsize=8)         # Set y-axis label for the third subplot with font size 8
axes[0, 2].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the third subplot with font size 8


axes[1, 0].plot(x, np.sqrt(x), 'y--')                       # Plot square root in the fourth subplot with yellow dashed line
axes[1, 0].set_title("Square Root", fontsize=10)            # Set title for the fourth subplot with font size 10
axes[1, 0].set_ylabel("Square Root Values", fontsize=8)     # Set y-axis label for the fourth subplot with font size 8
axes[1, 0].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the fourth subplot with font size 8


axes[1, 1].plot(x, np.log(x), 'm--')                        # Plot natural logarithm in the fifth subplot with magenta dashed line
axes[1, 1].set_title("Natural Logarithm", fontsize=10)      # Set title for the fifth subplot with font size 10
axes[1, 1].set_ylabel("Logarithm Values", fontsize=8)       # Set y-axis label for the fifth subplot with font size 8
axes[1, 1].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the fifth subplot with font size 8


axes[1, 2].plot(x, np.exp(x), 'c--')                        # Plot exponential in the sixth subplot with cyan dashed line
axes[1, 2].set_title("Exponential", fontsize=10)            # Set title for the sixth subplot with font size 10
axes[1, 2].set_ylabel("Exponential Values", fontsize=8)     # Set y-axis label for the sixth subplot with font size 8
axes[1, 2].set_xlabel("X Values", fontsize=8)               # Set x-axis label for the sixth subplot with font size 8


plt.tight_layout()                              # Adjust subplot parameters to give specified padding
plt.show()                                      # Display the plots
```

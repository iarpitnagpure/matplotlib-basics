```python
import matplotlib.pyplot as plt
import numpy as np

# Create two arrays using numpy's arange function
# arange(start, stop, step): Creates values from start to stop (exclusive) with a given step size
data1 = np.arange(0.0, 5.0, 0.1)
print(data1)
data2 = np.arange(0.0, 5.0, 0.2)
print(data2)

# Create a new figure for plotting
plt.figure()

# Create subplots and plot the data
# subplot(nrows, ncols, index): Creates a subplot in a grid of nrows x ncols at position index
plt.subplot(2, 1, 1)                    # Create a subplot (2 rows, 1 column, 1st subplot)
plt.plot(data1, data1, 'r--')           # Red dashed line for data1

plt.subplot(2, 1, 2)                    # Create a subplot (2 rows, 1 column, 2nd subplot)
plt.plot(data2, data2**2, 'bs')         # Blue squares for data2 squared

plt.show()                              # Display the plots
```

```python
# Create a new figure for plotting
plt.figure()

# Create subplots and plot the data
# subplot(nrows, ncols, index): Creates a subplot in a grid of nrows x ncols at position index
plt.subplot(1, 2, 1)                    # Create a subplot (1 row, 2 columns, 1st subplot)
plt.plot(data1, data1, 'r--')           # Red dashed line for data1

plt.subplot(1, 2, 2)                    # Create a subplot (1 row, 2 columns, 2nd subplot)
plt.plot(data2, data2**2, 'bs')         # Blue squares for data2 squared

plt.show()                              # Display the plots
```

```python
plt.figure()

plt.suptitle("Subplots Example")        # Add a main title for the entire figure

plt.subplot(1, 3, 1)                    # Create a subplot (1 row, 2 columns, 1st subplot)
plt.plot(data1, data1, 'r--')           # Red dashed line for data1
plt.xlabel('Data 1 X-axis Label')       # Label for X-axis
plt.ylabel('Data 1 Y-axis Label')       # Label for Y-axis

plt.subplot(1, 3, 2)                    # Create a subplot (1 row, 2 columns, 2nd subplot)
plt.plot(data2, data2**2, 'bs')         # Blue squares for data2 squared
plt.xlabel('Data 2 X-axis Label')       # Label for X-axis
plt.ylabel('Data 2 Y-axis Label')       # Label for Y-axis

plt.subplot(1, 3, 3)                    # Create a subplot (1 row, 2 columns, 3rd subplot)
plt.plot(data2, data2**3, 'g^')         # Green triangles for data2 cubed
plt.xlabel('Data 3 X-axis Label')       # Label for X-axis
plt.ylabel('Data 3 Y-axis Label')       # Label for Y-axis

plt.tight_layout()                      # Adjust layout to prevent overlap

plt.show()                              # Display the plots
```
